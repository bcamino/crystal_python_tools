### Position a nanoparticle on a surface

```python
import sys
sys.path.insert(1, '../functions')
from os.path import join
import os
import numpy as np

from ase.cluster import wulff_construction
from ase.visualize import view

from crystal_io import *

from settings import runcry
from geom_modification import make_supercell
from geom_modification import insert_atom
from geom_modification import displace_atom
from visualisation_tools import gui2cif
```

```python
#Variables used in the whole notebook
directory = '../data' # directory where the original input is saved
input_name = 'BN_5x5.d12' # name of the original input
np_input_name = 'BN_5x5_NP.d12'
file_path = join(directory,input_name)
np_file_path = join(directory,np_input_name)

#Supercell generation
```

```python
# Create the NP by using ASE
surfaces = [(1, 0, 0), (1, 1, 0), (1, 1, 1)]
esurf = [1.0, 1.1, 0.9]   # Surface energies.
lc = 3.61000
size =5  # Number of atoms
atoms = wulff_construction('Cu', surfaces, esurf,
                           size, 'fcc',
                           rounding='above', latticeconstant=lc)
n_atoms = len(atoms.get_atomic_numbers())
disp = False
view(atoms, viewer='x3d')
```

```python
# Read the NP structure
```

```python
# Move and rotate the NP
atoms.rotate(45,'x')
atoms.rotate(35,'y')
first_disp = 51
disp = np.array([4.372597334022, 2.51879531, 4.45])
```

```python
geom_block,optgeom_block,bs_block,func_block,scf_block = read_input(file_path)

geom_block = insert_atom(geom_block,atoms.get_atomic_numbers().tolist(),atoms.get_positions().tolist()) 

if type(disp) != bool:
    disp_atoms = [i for i in range(first_disp,(first_disp+n_atoms))]
    disp_coord = np.tile(disp, (n_atoms, 1)).tolist()
    geom_block = displace_atom(geom_block,disp_atoms,disp_coord) 
    
write_input(np_file_path,geom_block,bs_block,func_block,scf_block,optgeom_block) 

runcry(np_file_path[:-4])

gui2cif(np_file_path[:-4]+'.gui')
cif_file_name = np_file_path[:-4]+'.cif'

! /Applications/VESTA/VESTA.app/Contents/MacOS/VESTA $cif_file_name
```
